```python
def generateTable(data, k=4): #State size is a number of words the probability of a next word depends on
    table_markov = {}
    for i in range(1, k + 1):
        text_model = markovify.Text(data, state_size=i)#Markovify.Text uses a markov chain to create a unique sentence.
        for key in text_model.chain.model.keys():
            table_markov[' '.join(key)] = text_model.chain.model[key]
    return table_markov
```

```python
def Cleany(x):
    for p in x:
        if p in ['.', '?', '!']:
            x1 = x.replace(p, '')
    return x1
```

```python
def Cleangenerated(t):
    t1 = {}
    for x in t.keys():
        if '.' in x or '?' in x or '!' in x:
            x1 = Cleany(x)
            if (x1 in t.keys()):
                t1[x1] = t[x1].update(t[x])
            else:
                t1[x1] = t[x]
        else:
            t1[x] = t[x]
    t2 = {}
    for x, y in t1.items():
        t2[x] = {}
        if (type(y) == dict):
            for z in list(y):
                if '.' in z or '?' in z or '!' in z:
                    z1 = Cleany(z)
                    if (z1 in list(y)):
                        t2[x][z1] = t1[x][z] + t1[x][z1]
                    else:
                        t2[x][z1] = t1[x][z]
                else:
                    t2[x] = t1[x]
    return t2
```

```python
def convertFreqIntoProb(T):
    for kx in T.keys():
        s = float(sum(T[kx].values()))
        for k in T[kx].keys():
            T[kx][k] = T[kx][k] / s #probablity

    return T
```

```python
def MarkovChain(text, k=1):
    T = generateTable(text, k)
    T = convertFreqIntoProb(T)
    return T
    
```
